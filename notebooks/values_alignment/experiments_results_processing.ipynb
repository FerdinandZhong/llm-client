```python
import pandas as pd
from llm_client.pipeline import Pipeline
import time
from llm_client.values_alignment.experiments_utils import *
import os
```

```python
root_path = f"{os.environ['HOME']}/Projects/llm_client/notebooks/values_alignment/experiments_results" + "/{model}/{lang}"

print(root_path)
```

```python
question_cols = [f"m_{i+1}" for i in range(24)]

print(question_cols)
```

```python
vsm_cols = ["PDI", "IDV", "MAS", "UAI", "LTO", "IVR"]
```

# Single Model -- Single Language

## Llama2-7b

### No Shuffle

```python
llama2_7b_seed_no_shuffle_dfs = []

for seed in range(10):
    seed_df = pd.read_csv(f"{root_path.format(model='Llama2_7b', lang='english')}/no_shuffle/result_{seed}.csv", index_col=0)
    llama2_7b_seed_no_shuffle_dfs.append(seed_df)

llama2_7b_seed_no_shuffle_dfs[0].head()
```

```python
llama2_7b_score_dfs = retrieve_score(llama2_7b_seed_no_shuffle_dfs, "gender", "age", "nation", "city")
```

```python
llama2_7b_avg_score_df = get_avg_score_df(llama2_7b_score_dfs)

llama2_7b_avg_score_df.head()
```

```python
llama2_7b_vsm_score_df = get_vsm_score_df(llama2_7b_score_dfs)

llama2_7b_vsm_score_df.head()
```

```python
llama2_7b_nation_correlation_df, llama2_7b_nation_score, llama2_7b_nation_p_values = get_original_score_pearson_values(llama2_7b_avg_score_df, target_cols=question_cols, merge_by="nation")

print(llama2_7b_nation_score, llama2_7b_nation_p_values)
llama2_7b_nation_correlation_df.head()
```

```python
llama2_7b_gender_correlation_df, llama2_7b_gender_score, llama2_7b_gender_p_values = get_original_score_pearson_values(llama2_7b_avg_score_df, target_cols=question_cols, merge_by="gender")

print(llama2_7b_gender_score, llama2_7b_gender_p_values)
llama2_7b_gender_correlation_df.head()
```

```python
llama2_7b_age_correlation_df, llama2_7b_age_score, llama2_7b_age_p_values = get_original_score_pearson_values(llama2_7b_avg_score_df, target_cols=question_cols, merge_by="age")

print(llama2_7b_age_score, llama2_7b_age_p_values)
llama2_7b_age_correlation_df.head()
```

```python
llama2_7b_std = get_standard_varaince_over_df(llama2_7b_vsm_score_df, target_cols=vsm_cols, merge_by="nation")

print(llama2_7b_std)
```

### Shuffle

```python
llama2_7b_seed_shuffle_dfs = []

for seed in range(10):
    seed_df = pd.read_csv(f"{root_path.format(model='Llama2_7b', lang='english')}/shuffle/result_{seed}.csv", index_col=0)
    llama2_7b_seed_shuffle_dfs.append(seed_df)

llama2_7b_seed_shuffle_dfs[0].head()
```

```python
llama2_7b_shuffle_score_dfs = retrieve_score(llama2_7b_seed_shuffle_dfs, "gender", "age", "nation", "city")
```

```python
llama2_7b_shuffle_vsm_score_df = get_vsm_score_df(llama2_7b_shuffle_score_dfs)

llama2_7b_shuffle_vsm_score_df.head()
```

```python
llama2_7b_shuffle_std = get_standard_varaince_over_df(llama2_7b_shuffle_vsm_score_df, target_cols=vsm_cols, merge_by="nation")

print(llama2_7b_shuffle_std)
```

```python
llama2_7b_shuffle_distance_df, llama2_7b_shuffle_overall_distance = compute_vsm_values_gap(
    llama2_7b_vsm_score_df, 
    llama2_7b_shuffle_vsm_score_df,
    target_cols=vsm_cols, 
    merge_by="nation"
)
print(llama2_7b_shuffle_overall_distance)
llama2_7b_shuffle_distance_df.head()
```
